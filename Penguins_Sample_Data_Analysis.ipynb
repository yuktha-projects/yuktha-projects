**Lets find the average measurement per species**
Create a table to get an average mesurement of bill length, depth and flipper lenght


```python
avgSpeciesMeasurement = penguins.groupby('species')[['bill_length_mm', 'bill_depth_mm' , 'flipper_length_mm', 'body_mass_g']].mean()
avgSpeciesMeasurement
```

**Insights:**
* On an Average Chinstrap species are having bigger bill length and depth compared to other 2 species.
* on an average Gentoo species are having bigger flipper length and body mass

**Create a column (Size Category)**

* Classify the group into 3 categories
       

        *   Light weight: < 4000 g
        *   Medium weight: `4000-5000 g`
        *   Heavy Weight: > 5000 g


        

     

```python
penguins['sizeCategory'] = np.where(
    penguins['body_mass_g'] < 4000 , 'LightWeight', 'Heavy'
)
penguins
```

```python
penguins['sizeCategory'] = np.where(
    penguins['body_mass_g'] < 4000 , 'LightWeight',
    np.where(penguins['body_mass_g'] <= 5000, 'Medium', 'Heavy')
)
penguins
```

```python
penguins[['species','body_mass_g','sizeCategory']]
```
