**Bitcoin Price Trend + Moving Averages Dashboard**

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Loading Bitcoin price data (last 2 years)
data = yf.download("BTC-USD", period="2y", interval="1d")

# Calculating Moving Averages
data['MA20'] = data['Close'].rolling(20).mean()
data['MA50'] = data['Close'].rolling(50).mean()

# Droping rows where moving averages are not available yet
data = data.dropna()

# Ploting Price + Moving Averages
plt.figure(figsize=(14,6))
plt.plot(data['Close'], label='Bitcoin Closing Price')
plt.plot(data['MA20'], label='20-Day Moving Average')
plt.plot(data['MA50'], label='50-Day Moving Average')
plt.title('Bitcoin Price Trend with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()

data.tail()
```
